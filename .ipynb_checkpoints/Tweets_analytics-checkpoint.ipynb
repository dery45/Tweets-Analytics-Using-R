## Network Diagram For Topics

```r
myStopwords <- read_lines("data\\indonesian+english+twitter-stopwords-complete.txt")
```

```r
data_bigram <- tamsis
data_bigram
```

```r
data_bigram$full_text <- as.vector(data_bigram$full_text)
```

```r
bi.gram.words <- data_bigram %>% 
  unnest_tokens(
    input = full_text, 
    output = bigram, 
    token = 'ngrams', 
    n = 2
  ) %>% 
  filter(! is.na(bigram))
```

```r
bi.gram.words %>% 
  select(bigram) %>% 
  head(10)
```

```r
bi.gram.words %<>% 
  separate(col = bigram, into = c('word1', 'word2'), sep = ' ') %>% 
  filter(! word1 %in% myStopwords) %>% 
  filter(! word2 %in% myStopwords) %>% 
  filter(! is.na(word1)) %>% 
  filter(! is.na(word2)) %>% 
  filter(! grepl("\\d", word1)) %>% 
  filter(! grepl("\\d", word2))
```

```r
bi.gram.words %>% 
  head(10)
```

```r
bi.gram.count <- as.data.frame(table(bi.gram.words$word1, bi.gram.words$word2))
colnames(bi.gram.count) <- c("word1", "word2", "weight")
bi.gram.count <- bi.gram.count[order(bi.gram.count$weight, decreasing = TRUE), ]
```

```r
bi.gram.count %>% head()

bi.gram.count %>% 
  ggplot(mapping = aes(x = weight)) +
  theme_light() +
  geom_histogram() +
  labs(title = "Bigram Weight Distribution")
```

```r
bi.gram.count %>% 
  mutate(weight = log(weight + 1)) %>% 
  ggplot(mapping = aes(x = weight)) +
  theme_light() +
  geom_histogram() +
  labs(title = "Bigram log-Weight Distribution")
```

```r
threshold <- 18

# For visualization purposes we scale by a global factor. 
ScaleWeight <- function(x, lambda) {
  x / lambda
}
```

```r
network <-  bi.gram.count %>%
  filter(weight > threshold) %>%
  mutate(weight = ScaleWeight(x = weight, lambda = 2E3)) %>% 
  graph_from_data_frame(directed = FALSE)


network
is.weighted(network)
```

```r
plot(
  network, 
  vertex.size = 2,
  vertex.label.color = "blue", 
  vertex.label.cex = 1.4, 
  vertex.label.dist = 2,
  edge.color = 'blue', 
  #sub = glue('Weight Threshold: {threshold}'), 
  alpha = 50
)
```

```r
# Store the degree.
V(network)$degree <- strength(graph = network)

# Compute the weight shares.
E(network)$width <- E(network)$weight/max(E(network)$weight)

# Set the plot size
options(repr.plot.width = 15, repr.plot.height = 12)


network_plot <- plot(
  network, 
  vertex.color = '#001A66',
  vertex.size = 3,
  vertex.label.color = '#001A66', 
  vertex.label.cex = 2.2, 
  vertex.label.dist = 1.6,
  edge.color = '#001A66',  # Change edge color to blue
  edge.width = 3 * E(network)$width,
  main = 'Bigram Count Network', 
  alpha = 50
)
```

```r
ggsave("binetwork_pendekar_chart.png", network_plot, dpi = 300)
```

```r
clusters(graph = network)


# Select biggest connected component.  
V(network)$cluster <- clusters(graph = network)$membership

cc.network <- induced_subgraph(
  graph = network,
  vids = which(V(network)$cluster == which.max(clusters(graph = network)$csize))
)

cc.network 
```

```r
# Store the degree.
V(cc.network)$degree <- strength(graph = cc.network)

# Compute the weight shares.
E(cc.network)$width <- E(cc.network)$weight/max(E(cc.network)$weight)

plot(
  cc.network, 
  vertex.color = '#001A66',
  # Scale node size by degree.
  vertex.size = 1,
  vertex.label.color = '#001A66', 
  vertex.label.cex = 2, 
  vertex.label.dist = 1,
  edge.color = '#001A66', 
  # Set edge width proportional to the weight relative value.
  edge.width = 3*E(cc.network)$width ,
  #sub = glue('Weiight Threshold: {threshold}'), 
  alpha = 50
)
```

```r

# Treshold
threshold <- 6

network <-  bi.gram.count %>%
  filter(weight > threshold) %>%
  graph_from_data_frame(directed = FALSE)

# Store the degree.
V(network)$degree <- strength(graph = network)
# Compute the weight shares.
E(network)$width <- E(network)$weight/max(E(network)$weight)

# Create networkD3 object.
network.D3 <- igraph_to_networkD3(g = network)
# Define node size.
network.D3$nodes %<>% mutate(Degree = (1E-2)*V(network)$degree)
# Degine color group (I will explore this feature later).
network.D3$nodes %<>% mutate(Group = 1)
# Define edges width. 
network.D3$links$Width <- 10*E(network)$width

forceNetwork(
  Links = network.D3$links, 
  Nodes = network.D3$nodes, 
  Source = 'source', 
  Target = 'target',
  NodeID = 'name',
  Group = 'Group', 
  opacity = 0.9,
  Value = 'Width',
  Nodesize = 'Degree', 
  # We input a JavaScript function.
  linkWidth = JS("function(d) { return Math.sqrt(d.value); }"), 
  fontSize =40,
  zoom = TRUE, 
  opacityNoHover = 1
)
```

## Network Diagram for influencer

```r
data <- tamsis
```

```r
# Create an empty graph
g <- graph.empty()
```

```r
# Add vertices (users) to the graph
vertices <- unique(data$usernames)
g <- add.vertices(g, nv = length(vertices), name = vertices)
```

```r
# Add edges (connections) between users
edges <- data.frame(from = data$usernames[-1], to = data$usernames[-length(data$usernames)])
g <- add.edges(g, edges = as.matrix(edges))
```

```r
# Compute the centrality measures
centrality <- degree(g, mode = "in")
```

```r
# Get the top influencers based on indegree centrality
top_influencers <- head(sort(centrality, decreasing = TRUE), 15)
head(top_influencers, 10)
```

```r
# Create a subgraph with only the top influencers
subgraph <- induced.subgraph(g, vids = names(top_influencers))

# Compute the layout for the subgraph
layout <- layout_with_fr(subgraph)

# Set the plot size
options(repr.plot.width = 15, repr.plot.height = 12)

# Plot the network graph with customized settings
plot(
  subgraph,
  layout = layout,
  vertex.size = 3,  # Change vertex size to 3
  vertex.color = '#B3D9FF',  # Change vertex color to blue
  vertex.label.color = '#001A66',
  vertex.label.cex = 2,
  vertex.label.dist = 1.6,
  edge.color = '#B3D9FF',
  edge.width = 3 * E(subgraph)$width,  # Change edge width calculation
  main = 'Bigram Count Network',
  alpha = 50
)
```

## WordCloud 

```r
data <- tamsis
tweets <- data$full_text
tweets
```

```r
myStopwords <- read_lines("data\\indonesian+english+twitter-stopwords-complete.txt")
stopwords_id <- c(myStopwords, c("tamsis"))
```

```r
myCorpus <- Corpus(VectorSource(tweets))
myCorpus
```

```r
# convert to lower case
myCorpus <- tm_map(myCorpus, content_transformer(tolower))

# remove link
removeURL <- function(x) gsub("http[^[:space:]]*", "", x)
myCorpus <- tm_map(myCorpus, content_transformer(removeURL))

# remove anything other than English letters or space 
removeNumPunct <- function(x) gsub("[^[:alpha:][:space:]]*", "", x) 
myCorpus <- tm_map(myCorpus, content_transformer(removeNumPunct))
```

```r
#Remove using stopwords
myCorpus <- tm_map(myCorpus, removeWords, stopwords_id)
```

```r
# remove extra whitespace
myCorpus <- tm_map(myCorpus, stripWhitespace)

# keep a copy for stem completion later
myCorpusCopy <- myCorpus
```

```r
tdm <- TermDocumentMatrix(myCorpus, control = list(wordLengths = c(1, Inf)))
tdm
```

```r
#find word freq
freq.terms <- findFreqTerms(tdm, lowfreq = 7)
freq.terms[1:50]
```

```r
#grafik
term.freq <- rowSums(as.matrix(tdm))
term.freq <- subset(term.freq, term.freq >= 1000)
df <- data.frame(term = names(term.freq), freq = term.freq)
df
```

```r
ggplot2::ggplot(df, aes(x=term, y=freq)) + geom_bar(stat="identity") +
  xlab("Terms") + ylab("Count") + coord_flip() +
  theme(axis.text=element_text(size=7))
```

```r
m <- as.matrix(tdm)
m
```

```r
# calculate the frequency of words and sort it by frequency 
word.freq <- sort(rowSums(m), decreasing = T)
word.freq
```

```r
#Wordcloud
d <- data.frame(word = names(word.freq),freq=word.freq)
a <- head(d,n=100)

# Create wordcloud with color customization
wordcloud <- wordcloud2(a, shape = "cloud", backgroundColor = "white", color = "#001A66", size = 1)

# Display the wordcloud
wordcloud
```
